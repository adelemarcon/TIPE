```python
import numpy as np
from matplotlib import pyplot as plt
```

on fixe Tf = 300 K
on fait varier ZpnTm (0.1,05,1,2,5)

```python
def rendement(c):
    t = [i for i in range (300,1000)]
    n = []
    for i in t:
        x = ((i - 300)*((1+c)**(1/2) - 1))/(i*((1+c)**(1/2) + (300/i)))
        n.append(x)
    return(t,n,)

t01, n01 = rendement(0.1)
t05, n05 = rendement(0.5)
t1, n1 = rendement(1)
t2, n2 = rendement(2)
t5, n5 = rendement(5)
```

```python
plt.plot(t01,n01, label = "0.1")
plt.plot(t05,n05, label = "0.5")
plt.plot(t1,n1, label = '1')
plt.plot(t2,n2, label = "2")
plt.plot(t5,n5, label = "5")
plt.legend()
plt.grid()
plt.show()
```

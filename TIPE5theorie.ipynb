# 5. Rendement th√©orique

```python
import numpy as np
from matplotlib import pyplot as plt
```

on fixe Tf = 300 K
on fait varier ZpnTm (0.1,05,1,2,5)

```python
def rendement(c):
    t = [i for i in range (300,500)]
    n = []
    for i in t:
        Tm = (i + 300)/2
        x = ((i - 300)*((1+c*Tm)**(1/2) - 1))/(i*((1+c*Tm)**(1/2) + (300/i)))
        n.append(x)
    return(t,n,)

t01, n01 = rendement(0.001)
t05, n05 = rendement(0.01)
t1, n1 = rendement(0.1)
t2, n2 = rendement(1)
t5, n5 = rendement(10)
```

```python
plt.scatter(332.95, 0.068, color = 'black',zorder = 6, s = 20, label = "Exp√©rimental")
plt.scatter(319.75, 0.065, color = 'black', zorder = 6, s = 20)
plt.scatter(330.25, 0.1, color = 'black', zorder = 6, s = 20)
plt.scatter(336.95, 0.095, color = 'black', zorder = 6, s = 20)
plt.plot(t01,n01, label = "Zpn = 0.001", color = "#cd2d2d")
plt.plot(t05,n05, label = "Zpn = 0.01", color = "#f2802e")
plt.plot(t1,n1, label = 'Zpn = 0.1', color = "#7bba3e")
plt.plot(t2,n2, label = "Zpn = 1", color = "#54c8cf")
plt.plot(t5,n5, label = "Zpn = 10", color = "#b654cf")
plt.legend()
plt.xlabel("Tc (K)")
plt.ylabel("Œ∑max")
plt.title("Rendement")
plt.grid()
plt.show()
```

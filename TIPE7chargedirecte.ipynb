## Charge directe

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def lecture2(fichier):
    a = []
    b = []
    c = []
    with open(fichier,'r') as f:
        for ligne in f:
            ai,bi,ci = [float(val.strip()) for val in ligne.split('\t')]
            a.append(ai)
            b.append(bi)
            c.append(ci)
    return a, b, c
s,t,u = lecture2("TIPE7mesure00.csv")
t00 = np.array(s)
exp00 = np.array(t)
moy00 = np.array(u)


def ajustement2(fichier):
    t, exp, moy = lecture2(fichier)
    N = []
    T = []
    for i in range (len(t)):
        if np.abs(exp[i]-moy[i]) < 0.04:
            N.append(exp[i])
            T.append(t[i])
    return T, N
T00, A00 = ajustement2("TIPE7mesure00.csv")
```

```python
plt.scatter(t00,exp00, color = "#ff7dd3", marker = '.')
plt.scatter(t00,moy00, color = "#5260e9", marker = '.')
plt.scatter(T00, A00, color = '#b67dff')
plt.xlabel("t (s)")
plt.ylabel("Tension (V)")
plt.title("Charge directe d'une batterie")
plt.grid()
plt.show()
```

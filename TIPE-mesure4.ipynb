```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
Th = np.array([]) # à faire varier avec le bain thermostaté
Tc = np.array([]) # à mesurer
DeltaT = Th - Tc
U = np.array([]) # Tension mesurée
I = np.array([])
P = U*I


plt.plot(DeltaT,P,color = 'blue')
plt.scatter(DeltaT,P, color ="red")
plt.xlabel("DeltaT")
plt.ylabel("Puissance fournie par les cellules")
plt.title("Octogone")
plt.grid()
plt.legend()
plt.show()
```

## Évolution du nombre de cellules en dérivation

1. Température attendue à la surface de la cellule

```python
import numpy as np
from matplotlib import pyplot as plt
```

Expérience pour montrer l'effet du nombre de cellule ajoutées en parallèle sur la température

```python
Th = np.array([])
Tc = np.array([])
DeltaT = Th - Tc
Nb = np.array([1,2,3,4,5,6])

plt.plot(Nb,DeltaT,color = 'blue',label = "Tension 1")
plt.scatter(Nb,DeltaT, color ="red")
plt.xlabel("Nombre de cellules")
plt.ylabel("Delta T")
plt.title("Évolution du Delta T selon le nombre de cellules")
plt.grid()
plt.legend()
plt.show()
```

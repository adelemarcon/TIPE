## Détermination de la résistance interne

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def lecture(fichier):
    t = []
    a = []
    with open(fichier,'r') as f:
        for ligne in f:
            ti,ai = [float(val.strip()) for val in ligne.split('\t')]
            t.append(ti)
            a.append(ai)
    return t, a
a,b = lecture("TIPE4mesure47.csv")
R47 = np.array(a)
P47 = np.array(b)
c,d = lecture("TIPE4mesure57.csv")
R57 = np.array(c)
P57 = np.array(d)
e,f = lecture("TIPE4mesure64.csv")
R64 = np.array(e)
P64 = np.array(f)

def theorique(R,T,r):
    alpha = 0.0425
    theo = []
    for i in range (len(R)):
        c = (R[i]*alpha*alpha*T*T)/((R[i]+r)**2)
        theo.append(c)
    return theo

Ptheo47 = np.array(theorique(a,46.6,3.4))
Ptheo57 = np.array(theorique(c,57.1,3.6))
Ptheo64 = np.array(theorique(e,63.8,3.5))
```

```python
plt.plot(R47,Ptheo47,color ="#00CCFF")
plt.plot(R57,Ptheo57, color = "#6600FF")
plt.plot(R64,Ptheo64, color = "#FF00FF")
plt.scatter(R47, P47, color ="#99FFFF", label = "ΔT = 46,6°C", marker = '+')
plt.scatter(R57,P57, color = "#9999FF", label = "ΔT = 57.1°C", marker = '+')
plt.scatter(R64,P64, color = "#FFCCFF", label = "ΔT = 63.8°C", marker = '+')
plt.xlabel("R (Ohm)")
plt.ylabel("Puissance (W)")
plt.title("Détermination de la résistance interne")
plt.grid()
plt.legend()
plt.show()
```

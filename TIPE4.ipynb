## Phénomène de dérive

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def lecture(fichier):
    t = []
    a = []
    with open(fichier,'r') as f:
        for ligne in f:
            ti,ai = [int(float(val.strip())) for val in ligne.split('\t')]
            t.append(ti)
            a.append(ai)
    return t, a
a,b = lecture("TIPE4mesure47.csv")
R = np.array(a)
P47 = np.array(b)
```

```python
plt.plot(R,P47, color ="pink")
plt.xlabel("R (Ohm)")
plt.ylabel("Puissance (W)")
plt.title("Détermination de la résistance interne")
plt.grid()
plt.show()
```

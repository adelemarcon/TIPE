## Octogone: Tension en fonction du delta T

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def lecture(fichier):
    a = []
    b = []
    with open(fichier,'r') as f:
        for ligne in f:
            ai,bi = [float(val.strip()) for val in ligne.split('\t')]
            a.append(ai)
            b.append(bi)
    return a, b

DT1,U1 = lecture("TIPE2octogonemesure.csv")
DT2,U2 = lecture("TIPE2octogonemesure2.csv")

plt.plot(DT1,U1,color = 'blue', marker = '.', label = 'Thermomètre sur cellules')
plt.plot(DT2,U2,color = 'red', marker = '.', label = "Thermomètre dans liquides")
plt.xlabel("DeltaT (°C)")
plt.ylabel("Tension en circuit ouvert (V)")
plt.title("Mise en évidence du phénomène")
plt.legend()
plt.grid()
plt.show()
```

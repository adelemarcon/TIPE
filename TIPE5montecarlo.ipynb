# Monte-Carlo

```python
import numpy as np
from numpy import random
from matplotlib import pyplot as plt

# Mesures et demi étendues associées (hypothèse de distributions uniformes)
# -------------------------------------------------------------------------
Pr = 0.5
dPr = 0.01
Pf = 6.25
dPf = 0.05

# Simulation de Monte Carlo
# -------------------------
N = 1000
Prmc= random.uniform (Pr-dPr,Pr + dPr,N)
Pfmc = random.uniform (Pf-dPf,Pf + dPf,N )
nmc = Prmc / Pfmc

# Présentation des résultats
# -------------------------- 
plt.subplot(1,3,1)
plt.hist(Prmc,bins='rice',color='red')
plt.xlabel('petit côté (cm)')
plt.ylabel('effectifs')

plt.subplot(1,3,2)
plt.hist(Pfmc,bins='rice',color='orange')
plt.xlabel('grand côté (cm)')

plt.subplot(1,3,3)
plt.hist (nmc, bins='rice',color='green')
plt.xlabel('aire feuille A4 ($cm^2$)')

print ("rendement : {:.2f} %".format(np.mean(nmc*100)))
print ("Incertitude-type : {:.2f} %".format(np.std(nmc*100,ddof=1)))
```

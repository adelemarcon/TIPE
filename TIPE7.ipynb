## Recharge d'une batterie

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def lecture(fichier):
    a = []
    b = []
    c = []
    with open(fichier,'r') as f:
        for ligne in f:
            ai,bi,ci = [float(val.strip()) for val in ligne.split(';')]
            a.append(ai)
            b.append(bi)
            c.append(ci)
    return a, b, c

def lecture2(fichier):
    a = []
    b = []
    c = []
    with open(fichier,'r') as f:
        for ligne in f:
            ai,bi,ci = [float(val.strip()) for val in ligne.split('\t')]
            a.append(ai)
            b.append(bi)
            c.append(ci)
    return a, b, c
a,b,c = lecture("TIPE7mesure07.csv")
t07 = np.array(a)
exp07 = np.array(b)
moy07 = np.array(c)
d,e,f = lecture("TIPE7mesure075.csv")
t075 = np.array(d)
exp075 = np.array(e)
moy075 = np.array(f)
g,h,i = lecture("TIPE7mesure08(1).csv")
t081 = np.array(g)
exp081 = np.array(h)
moy081 = np.array(i)
j,k,l = lecture("TIPE7mesure085.csv")
t085 = np.array(j)
exp085 = np.array(k)
moy085 = np.array(l)
m,n,o = lecture("TIPE7mesure09.csv")
t09 = np.array(m)
exp09 = np.array(n)
moy09 = np.array(o)
p,q,r = lecture2("TIPE7mesure08(2).csv")
t082 = np.array(p)
exp082 = np.array(q)
moy082 = np.array(r)
s,t,u = lecture2("TIPE7mesure00.csv")
t00 = np.array(s)
exp00 = np.array(t)
moy00 = np.array(u)

```

```python

plt.scatter(t07, moy07, color ="red", label = "α = 0.7", marker = '.')
plt.scatter(t075,moy075, color = "orange", label = "α = 0.75", marker = '.')
plt.scatter(t081,moy081, color = "yellow", label = "α = 0.8", marker = '.')
plt.scatter(t085,moy085, color = "blue", label = "α = 0.85", marker = '.')
plt.scatter(t09,moy09, color = "purple", label = "α = 0.9", marker = '.')
plt.scatter(t082,moy082, color = "green", label = "α = 0.8", marker = '.')
plt.xlabel("t (s)")
plt.ylabel("Tension (V)")
plt.title("Recharge d'une batterie pour différents coefficients α")
plt.grid()
plt.legend()
plt.show()
```

## Phénomène de dérive

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
t1 = np.array([])
t2 = np.array([])
Temp1 = np.array([])
Temp2 = np.array([])

plt.plot(t1,Temp1,color = 'red',label = "Sans refroidissement")
plt.plot(t2,Temp2,color = 'blue',label = "Avec refroidissement")
plt.scatter(t1,Temp1, color ="black")
plt.scatter(t2,Temp2, color ="black")
plt.xlabel("Temps (s)")
plt.ylabel("Température")
plt.title("Évolution de la température au cours du temps")
plt.grid()
plt.legend()
plt.show()
```

## Charge optimale pour 0.8

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def lecture(fichier):
    a = []
    b = []
    c = []
    with open(fichier,'r') as f:
        for ligne in f:
            ai,bi,ci = [float(val.strip()) for val in ligne.split(';')]
            a.append(ai)
            b.append(bi)
            c.append(ci)
    return a, b, c

g,h,i = lecture("TIPE7mesure08(1).csv")
t081 = np.array(g)
exp081 = np.array(h)
moy081 = np.array(i)

def ajustement(fichier):
    t, exp, moy = lecture(fichier)
    N = []
    T = []
    for i in range (len(t)):
        if np.abs(exp[i]-moy[i]) < 0.03:
            N.append(exp[i])
            T.append(t[i])
    return T, N

T081,A081 = ajustement("TIPE7mesure08(1).csv")
```

```python
plt.scatter(t081,moy081, color = "#5260e9", label = "moyenne", marker = '.')
plt.scatter(t081,exp081, color = "#ff7dd3", label = "exp", marker = '.')
plt.scatter(T081, A081, color ="#b67dff", label = "exp ajustée", marker = '.')
plt.xlabel("t (s)")
plt.ylabel("Tension (V)")
plt.title("Recharge optimale d'une batterie")
plt.grid()
plt.legend()
plt.show()
```
